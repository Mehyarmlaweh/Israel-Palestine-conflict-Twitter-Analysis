```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('FinalDataSet.csv')
```

```python
sentiment_counts = df['Category'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Sentiment Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
support_counts = df['Support'].value_counts()

# Assuming 1 corresponds to 'Supporters of Palestine'
labels = ['Supporters of Palestine','Neutral', 'Opponents of Israel']

bars = plt.bar(labels, support_counts.values, color=['lightcoral', 'lightblue', 'lightgreen'])

for label, count, bar in zip(labels, support_counts.values, bars):
    if label == 'Supporters of Palestine':
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, str(count), ha='center', va='bottom')
    else:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, str(count), ha='center', va='top')

plt.xlabel('Supporters')
plt.ylabel('Count')
plt.title('Support Distribution')
plt.show()
```
